```python
#type: ignore

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
customers = pd.read_csv('Mall_Customers.csv')

#let's cluster by each individual's annual income and spending score
X = customers.iloc[:, -2:]

#finding the best number of clusters using elbow rule
inertias, results = [None, None], [None, None]

for k in range(2, 11):
    model = KMeans(n_clusters=k).fit(X)
    inertias.append(model.inertia_)
    results.append(silhouette_score(X, model.labels_))

sns.lineplot(inertias)
plt.title('inertias')
plt.show()
```

```python
sns.lineplot(results)
plt.title('scores')
plt.show()
```

```python
#by the elbow rule and evaluating quality of clustering, we choose K = 5 as the best number of centroids

model = KMeans(n_clusters = 5).fit(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(data = X, x = 'Annual Income (k$)', y = "Spending Score (1-100)", hue=model.labels_)
plt.show()
```
